```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("https://covidtracking.com/api/us/daily.csv")
```

```python
df["positive_test_rate"] = df.positive / df.totalTestResults
```

```python
def convert_date(date):
    date_string = str(date)
    return pd.Timestamp(int(date_string[:4]), int(date_string[4:6]), int(date_string[6:]))

df["datetime"] = df.date.apply(convert_date)
```

```python
g = sns.lineplot(data=df, x="datetime", y="positive_test_rate")
plt.xticks(rotation=45)
```
